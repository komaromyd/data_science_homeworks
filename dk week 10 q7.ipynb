### Principal components: eigenvalues of the covariance matrix, expressed as linear combinations of the original variables:
### 0. The dataset values are normalized and re-centered, so that the standard deviation is 1 and the mean is 0.
### 1. The covariance matrix is a matrix $M$ whose element $\ M_{ij} $ is the covariance of the $i^{th}$ and $j^{th}$  column of the matrix (the values of the the $i^{th}$ and $j^{th}$ variable).
### 2. Then, $M$ is diagonalized. The eigenvectors are the principal components, which are linear combinations of the original variables.
### 3. The sum of the eigenvalues is 1. If the sum of the first $k$ eigenvalues is $p$, then the first $k$ principal components express $p$ part of the overall variance. 

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_labels = breast.target
```

```python
breast_labels
```

```python
breast_labels.shape
```

```python
breast_data
```

```python
labels = np.reshape(breast_labels,(569,1))
```

```python
labels
```

```python
final_breast_data = np.concatenate([breast_data,labels],axis=1)
```

```python
final_breast_data
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset
```

```python
breast
```

```python
features = breast.feature_names
features
```

```python
features_labels = np.append(features,'label')
```

```python
features_labels
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset.tail()
```

```python
breast_dataset['label'].replace(0, 'Benign',inplace=True)
breast_dataset['label'].replace(1, 'Malignant',inplace=True)
```

```python
breast_dataset.tail()
```

```python
x = breast_dataset.loc[:, features].values
```

```python
x
```

```python
x.shape
```

```python
np.mean(x),np.std(x)
```

```python
x = StandardScaler().fit_transform(x) # normalizing the features
```

```python
x
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
feat_cols
```

```python
normalised_breast = pd.DataFrame(x,columns=feat_cols)
```

```python
normalised_breast
```

```python
pca_breast = PCA(n_components=2)
```

```python
pca_breast
```

```python
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principal_breast_Df
```

```python
print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
```

```python
print(pca_breast.explained_variance_ratio_)
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
