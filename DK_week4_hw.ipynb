```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
total_players=len(purchase_data.index)
dict0=[{"Total Players":total_players}]
df0 = pd.DataFrame(dict0)
df0
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items=len(purchase_data["Item Name"].unique())
av_price=purchase_data["Price"].mean()
nr_purch=len(purchase_data["Purchase ID"].unique())
tot_rev=purchase_data["Price"].sum()
dict1=[{"Number of Unique Items": unique_items, "Average Price": av_price, "Number of Purchases":nr_purch, "Total Revenue": tot_rev}]

new_df = pd.DataFrame(dict1)
new_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
members =[]
for gender in purchase_data["Gender"].unique():
    members.append(len(purchase_data.groupby("Gender")["SN"].value_counts().loc[gender]))
percentage=np.array(members)/sum(members)
array2=np.transpose(np.vstack([members, percentage]))
df2=pd.DataFrame(data=array2, index = ['Male', 'Other / Non-Disclosed', 'Female'], columns = ['Total Count', 'Percentage of Players']).round(decimals=2)
df2
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
df31=purchase_data.groupby("Gender").count()["Purchase ID"]
df32=purchase_data.groupby("Gender")["Price"].mean()
df33=purchase_data.groupby("Gender")["Price"].sum()
df3=pd.concat([df31, df32, df33], axis=1).round(decimals=2)
df3.columns=['Purchase Count', 'Average Purchase Price / $', 'Total Purchase Value / $']
df3
#missing: average purchase price per person #purchase_data.groupby(["Gender", "SN"])["Price"].sum()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
narrowed_df=purchase_data.loc[:, "SN": "Age"].drop_duplicates()
age_bins=[4,9,14,19,24,29,34,39,44]
age_labels=["10-", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(narrowed_df["Age"], age_bins, labels=age_labels)
narrowed_df["Age"]=pd.cut(narrowed_df["Age"], age_bins, labels=age_labels)
df41=narrowed_df.groupby("Age").count()["SN"]
df42=100*narrowed_df.groupby("Age").count()["SN"]/(narrowed_df.groupby("Age").count()["SN"].sum())
df4=pd.concat([df41, df42], axis=1).round(decimals=2)
df4.columns=['Total Count', 'Percentage of Players']
df4
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["agebins"]=pd.cut(purchase_data["Age"], age_bins, labels=age_labels)
df51=purchase_data["agebins"].value_counts().sort_index()
df52=purchase_data.groupby("agebins")["Price"].mean().round(decimals=2)
df53=purchase_data.groupby("agebins")["Price"].sum().round(decimals=2)
df5=pd.concat([df51, df52, df53], axis=1)
df5.columns=['Purchase Count', 'Average Purchase Price / $', 'Total Purchase Value / $']
#missing: average purchase per person
df5
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df61=purchase_data.groupby(["SN"]).count().sort_values("Price", ascending=False)["Price"].round(decimals=2)
df63=purchase_data.groupby(["SN"]).sum().sort_values("Price", ascending=False)["Price"].round(decimals=2)
df6=pd.concat([df61, df63], axis=1)
df6.columns=['Purchase Count', 'Total Purchase Value / $']
df6.sort_values("Total Purchase Value / $", ascending=False).head()
#missing: average purchase per person
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pd_item=purchase_data[["Item ID", "Item Name", "Price"]]
pd_item
df71=pd_item.groupby(["Item ID", "Item Name"]).mean().sort_values("Item ID", ascending= False)
df71.rename(columns = {'Price':'Item Price'}, inplace = True)
df72=pd_item.groupby(["Item ID", "Item Name"]).count().sort_values("Item ID", ascending= False)
df72.rename(columns = {'Price':'Purchase Count'}, inplace = True)
df73=pd_item.groupby(["Item ID", "Item Name"]).sum().sort_values("Item ID", ascending= False)
df73.rename(columns = {'Price':'Total Purchase Value'}, inplace = True)
df7raw=pd.concat([df71, df72, df73], axis=1)
df7=df7raw.sort_values( "Purchase Count", ascending = False)
df7
```
