```python
def revert(N):
    return(int(str(N)[::-1].lstrip('0')))
```

```python
revert(123456098790500)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plotrev(N):
    xlist=[]
    ylist=[]
    for j in range(1,N):
        xlist.append(j)
        ylist.append(revert(j))
    plt.scatter(xlist,ylist, s=2)
    plt.xscale("log")
    plt.yscale("log")
```

```python
plotrev(10000)
```
