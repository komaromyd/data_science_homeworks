```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import matplotlib.ticker as mticker
```

```python
coviddata=pd.read_csv('owid-covid-data.csv',parse_dates=True, index_col =["date"])
```

```python
covd_SWE=coviddata[coviddata["location"]=="Sweden"]
covd_FIN=coviddata[coviddata["location"]=="Finland"]
fig, ax = plt.subplots()
ax.plot(covd_SWE.index, covd_SWE["new_cases_smoothed"], color = "red")
ax.set_xlabel("Date") 
ax.set_ylabel("New daily cases (smoothed)", color = "red")
ax.tick_params('y', colors='red')
ax.xaxis.set_major_locator(mticker.MaxNLocator(6))
ticks_loc = ax.get_xticks().tolist()
ax.xaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
ax2 =ax.twinx()
ax2.plot(covd_FIN.index, covd_FIN["new_cases_smoothed"], color = "blue")
ax2.set_ylabel("New daily cases (smoothed)", color = "blue")
ax2.tick_params('y', colors='blue')
ax2.xaxis.set_major_locator(mticker.MaxNLocator(6))
ticks_loc = ax2.get_xticks().tolist()
ax2.xaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
plt.show()
```
